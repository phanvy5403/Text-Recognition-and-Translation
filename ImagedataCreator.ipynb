```python
import pandas as pd
import matplotlib.pyplot as plt
from PIL import Image, ImageFont, ImageDraw
labels = pd.read_csv('/content/data/Train/label_data.csv')
labels.head(8)
```

```python
#@title Converting Function here
import json

#value in csv cell is read as string, use this to convert to dictionary
def read_dic_cell(cell_value):
  return str(json.loads(cell_value)['text1'])

#return [x, y, width, height]
def read_pos(cell_value):
  out = json.loads(cell_value)
  return [out['x'], out['y'], out['width'], out['height']]

def size_to_coor(out):
  x, y, w, h= out
  x1, y1 = x,y
  x2, y2 = x1+w, y1+h
  return (x1, y1, x2, y2)
```

##Train dataset

```python
!touch '/content/data/Train_crop/label.txt'
```

```python

new_img_path = '/content/data/Train_crop/Images/'
img_path = '/content/data/Train/Images/'
file_label = open('/content/data/Train_crop/label.txt', 'a')
labels=pd.read_csv('/content/data/Train/label_data.csv')

img_id=0
for index, value in labels.iterrows():
  file_name = img_path+ value['filename']
  pil_img  = Image.open(file_name)

  file_name_new = new_img_path+ value['filename']
  #img_name =  id + '.jpg/.JPG/.png'
  img_name = str(img_id) + '.' + value['filename'].split('.')[1]

  #create label file
  file_label.write(new_img_path + img_name+'\t"'+ read_dic_cell(value['region_attributes']) + '"\n')

  #croping image
  rec = size_to_coor(read_pos(value['region_shape_attributes']))
  pil_img.crop(rec).save('/content/data/Train_crop/Images/'+ img_name)

  #new_file.loc[img_id] = [img_name, read_dic_cell(value['region_attributes'])]
  img_id += 1
```
